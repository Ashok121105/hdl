## Phase 1: Data Preprocessing & EDA

```python
print('Dataset shape:', df.shape)
print('\nInfo:')
df.info()
print('\nMissing values:')
print(df.isnull().sum())
print('\nDuplicates:', df.duplicated(subset=FEATURE_COLUMNS + [TARGET_COLUMN]).sum())
df.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[FEATURE_COLUMNS + [TARGET_COLUMN]].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, col in zip(axes.ravel(), FEATURE_COLUMNS):
    sns.scatterplot(data=df, x=col, y=TARGET_COLUMN, ax=ax, alpha=0.7)
    ax.set_title(f'{col} vs HDI')
plt.tight_layout()
plt.show()
```
